# Задание 1.

В re отстутсвуют позитивные квантификаторы. Добавление + к жадному квантификатору делает его позитивным. Он похож
на жадного, но без бэктрекинга. Основное практическое преимущество позитивных квантификаторов
состоит в том, что они ускоряют работу регулярного выражения. В частности позволяют регулярному выражению
"сломаться" быстрее.

```python
import regex
text = 'авбвг авбвбвг'

# Работает по-разному
result = regex.sub(r'а(б|в)*вг', '-', text)
print(result)
```

(б|в)*+ ищет 'б' или 'в', пока возможно, но без бэктрекинга, поэтому 'в' между ' * ' и 'г' никогда не будет найден

```python
result = regex.sub(r'а(б|в)*+вг', '-', text)
print(result)
```
